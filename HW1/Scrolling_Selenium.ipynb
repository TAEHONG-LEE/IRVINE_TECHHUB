```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
import time

# ChromeDriver를 자동으로 설치하고 실행합니다.
service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service)

collected_data = []  # 수집된 데이터를 저장할 리스트

# 2페이지와 3페이지의 데이터를 수집하기 위해 페이지 번호 설정
for page in range(3, 21):
    driver.get(f"https://www.yes24.com/24/Category/NewProductList/001?SumGb=04&FetchSize=80&PageNumber={page}")
    
    if page == 3:
        time.sleep(10)
    time.sleep(2)  # 페이지 로딩 대기

    # 페이지의 모든 책 이미지에 연결된 a 태그 찾기
    book_elements = driver.find_elements(By.CSS_SELECTOR, 'div.goodsImgW a')

    # a 태그의 href 속성 추출하여 리스트에 추가
    image_links = []
    for book in book_elements:
        href = book.get_attribute('href')
        if href and 'Goods' in href:  # 'Goods'를 포함하는 href만 수집
            image_links.append(href)

    # 각 링크를 순회하면서 데이터 수집
    for link in image_links:
        driver.get(link)
        
        try:
            adult_varification = driver.find_element(By.CLASS_NAME, 'servName')
            if adult_varification.text == '성인컨텐츠 본인인증':
                continue
        except Exception as e:
            pass
        WebDriverWait(driver, 2).until(EC.presence_of_element_located((By.CLASS_NAME, "gd_titArea")))

        # 필요한 정보를 수집합니다.
        try:
            title = driver.find_element(By.CSS_SELECTOR, 'h2.gd_name').text
            author = driver.find_element(By.CSS_SELECTOR, 'span.gd_auth').text
            publish_date = driver.find_element(By.CSS_SELECTOR, '.gd_date').text
            publisher = driver.find_element(By.CSS_SELECTOR, '.gd_pub').text
            original_price = driver.find_element(By.CLASS_NAME, 'yes_m').text
            sale_price = driver.find_element(By.CSS_SELECTOR, '.nor_price .yes_m').text
            
            book_info = {
                'title': title,
                'author': author,
                'publisher': publisher,
                'publish_date': publish_date,
                'original_price': original_price,
                'sale_price': sale_price,
            }

            collected_data.append(book_info)

        except Exception as e:
            print(f"An error occurred while processing the link {link}: {e}")

# 데이터 수집 완료 후 driver 종료
driver.quit()

# 수집된 데이터 확인
# print(collected_data)
```
